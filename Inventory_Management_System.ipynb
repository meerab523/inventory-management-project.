```python
 #******************************
# Name = Zainab Siddiqui.
# Roll no. = 381771.
# course = Python Programming.
#******************************

# <---------------------------------->
# <==================================>
#  SIMPLE INVENTORY MANAGEMENT SYSTEM.
# <==================================>
# <---------------------------------->

import json

def save_inventory():
    with open("inventory.json", "w") as file:
        json.dump(inventory, file, indent=4)

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +

def load_inventory():
    global inventory
    try:
        with open("inventory.json", "r") as file:
            inventory = json.load(file)
    except FileNotFoundError:
        inventory = []

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +

inventory = []
load_inventory()
if not inventory:
    inventory = [
        {"name": "Sugar", "quantity": 5, "price": 70.0, "category": "Grocery"},
        {"name": "Milk", "quantity": 10, "price": 50.0, "category": "Dairy"},
        {"name": "Soap", "quantity": 20, "price": 30.0, "category": "Toiletries"},
        {"name": "Eggs", "quantity": 12, "price": 180.0, "category": "Poultry"},
        {"name": "Rice", "quantity": 8, "price": 100.0, "category": "Grocery"},
    ]
    save_inventory()


# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def login():
  correct_username = "admin"
  correct_password = "123"
  tries = 3

  while tries > 0:
     username = input("üë§ Enter username: ").strip()
     password = input("üîë Enter password: ").strip()

     if correct_username == username and correct_password == password:
          print("‚úÖ  Access granted\n")
          main_menu()
          return
     else:
        tries -= 1
     if tries > 0:
        print(f"‚ùå Access denied. You have {tries} tries left.\n")
     else:
          print("üö´ Too many incorrect attempts. Exiting program.")
          exit()

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def add_item():
    name = input("Enter item name: ").strip().lower().title()
    for item in inventory:
        if item["name"] == name:
          print(f"\n‚ùå Item '{name.title()}' already exists in inventory.\n")
          return

    try:
      quantity = int(input("Enter quantity: "))
      price = float(input("Enter price: "))
    except ValueError:
      print("‚ùå Invalid quantity or price.")
      return

    category = input("Enter category: ").strip().lower().title()

    item = {
        "name": name,
        "quantity": quantity,
        "price": price,
        "category": category
    }
    inventory.append(item)
    save_inventory()
    print(f"\n‚úÖ Item '{name.title()}' added successfully.\n")

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def update_item():
    name = input("Enter the name of the item you want to update: ").strip().lower().title()
    for item in inventory:
        if item["name"].lower() == name.lower():
            print(f"\nCurrent details of '{name.title()}':")
            print(f"Quantity: {item['quantity']}, Price: {item['price']}, Category: {item['category'].title()}")
            print("-" * 40)

            # Ask what user wants to update
            new_quantity = (input("Enter new quantity (press Enter to skip): ")).strip()
            new_price = (input("Enter new price (press Enter to skip): ")).strip()
            new_category = input("Enter new category (press Enter to skip): ").strip().lower().title()

            # Update only if user entered new value
            if new_quantity:
              try:
                item["quantity"] = int(new_quantity)
              except ValueError:
                print("‚ùå Invalid quantity. Update skipped.")
            if new_price:
              try:
                item["price"] = float(new_price)
              except ValueError:
                print("‚ùå Invalid price. Update skipped.")
            if new_category:
                item["category"] = new_category

            print(f"\n‚úÖ '{name.title()}' updated successfully!\n")
            save_inventory()
            return
    print(f"\n‚ùå Item '{name.title()}' not found in inventory.\n")

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def delete_item():
    product_name = input("üóëÔ∏è Enter the name of the product to delete: ").strip().lower().title()
    for item in inventory:
        if item["name"].lower() == product_name.lower():
            confirm = input(f"Are you sure you want to delete '{product_name}'? (yes/no): ").strip().lower()
            if confirm != 'yes':
                print("‚ùå Deletion cancelled.")
                return

            inventory.remove(item)
            print("‚úÖ Item deleted successfully!\n")
            break
    else:
        print("‚ùå Product not found in inventory.\n")
    save_inventory()

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def search_item():
    product_name = input("üîç Enter the name of the product to search: ").strip().lower().title()
    for item in inventory:
        if item["name"].lower() == product_name.lower():
            print("\nüì¶ Product Found:")
            print(f"üõçÔ∏è Name: {item['name'].title()}")
            print(f"üî¢ Quantity: {item['quantity']}")
            print(f"üí∞ Price: {item['price']}")
            print(f"üìÇ Category: {item['category'].title()}\n")
            break
    else:
        print("‚ùå Product not found.\n")

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def display_inventory():
    if len(inventory)==0:
        print("üì≠ Inventory is empty.\n")
    else:
        print("üìã Current Inventory:")
        for item in inventory:
            print("-----------------------------")
            print(f"üõçÔ∏è Name     : {item['name'].title()}")
            print(f"üî¢ Quantity : {item['quantity']}")
            print(f"üí∞ Price    : {item['price']}")
            print(f"üìÇ Category : {item['category'].title()}")
        print("-----------------------------\n")

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def calculate_total_worth():
  if len(inventory)==0:
    print("Inventory is empty!\n")
  else:
    total=0
    for item in inventory:
      total += item["price"] * item["quantity"]
    print("üí∞ Total Inventory Worth: Rs.", total)
  save_inventory()

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def filter_by_category():
    if not inventory:
        print("Inventory is empty.")
        return

    # Make a list of all unique categories
    categories = []
    for item in inventory:
        cat = item["category"].strip().title()
        if cat not in categories:
            categories.append(cat)

    # Show all available categories
    print("\nAvailable Categories:")
    for index, category in enumerate(categories):
        print(f"{index + 1}. {category}")

    # Ask user to choose a category
    choice = input("Enter category number: ").strip()

    if not choice.isdigit() or int(choice) < 1 or int(choice) > len(categories):
        print("Invalid choice.")
        return

    selected_category = categories[int(choice) - 1]

    # Show items from the selected category
    print(f"\nItems in category: {selected_category.title()}")
    for item in inventory:
        if item["category"].strip().title() == selected_category:
            print(f"Name: {item['name'].title()}, Price: {item['price']}, Quantity: {item['quantity']}")
    save_inventory()

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def low_stock_alert():
    if not inventory:
        print("Inventory is empty.")
        return

    try:
        threshold = int(input("Enter the quantity threshold (default is 5): ") or "5")
    except ValueError:
        print("‚ùå Invalid input. Showing items with quantity < 5.")
        threshold = 5

    found = False
    print(f"\nItems with quantity less than {threshold}:\n")
    for item in inventory:
        if item["quantity"] < threshold:
            print(f"Name: {item['name'].title()}, Quantity: {item['quantity']}, Category: {item['category'].title()}")
            found = True

    if not found:
        print("‚úÖ All items are sufficiently stocked.")
    save_inventory()

# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
def restock_item():
    if not inventory:
        print("üì¶ Inventory is empty.")
        return

    name = input("Enter item name to restock: ").strip().lower().title()
    for item in inventory:
        if item["name"].lower() == name.lower():
          while True:
            try:
                add_qty = int(input("Enter quantity to add: "))
                if add_qty <= 0:
                    print("‚ùå Quantity must be positive.")
                    return
                item["quantity"] += add_qty
                print(f"‚úÖ '{name.title()}' restocked successfully. New quantity: {item['quantity']}")
                save_inventory()
                return
            except ValueError:
                print("‚ùå Invalid quantity entered.")
          return

    print(f"‚ùå Item '{name.title()}' not found in inventory.")
    save_inventory()


# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- #
# ------------------------------------------+ SIMPLE MENUE TO TEST THE FUNCTIONS. +-------------------------------------- #
# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- #

def main_menu():
    while True:
        print("\n" + "="*50)
        print("      üì¶ WELCOME TO INVENTORY MANAGEMENT SYSTEM")
        print("="*50)
        print("1. Add Item")
        print("2. Update Item")
        print("3. Delete Item")
        print("4. Search Item")
        print("5. Display Inventory")
        print("6. Total Inventory Worth")
        print("7. Filter by Category")
        print("8. Low Stock Alert")
        print("9. Restock Item")
        print("10. Exit")
        print("="*50)

        try:
          choice = input("Enter your choice (1-10): ").strip()
        except KeyboardInterrupt:
          print("\n Program interrupted. Good bye!")
          break

        if choice == "1":
            add_item()
        elif choice == "2":
            update_item()
        elif choice == "3":
            delete_item()
        elif choice == "4":
            search_item()
        elif choice == "5":
            display_inventory()
        elif choice == "6":
            calculate_total_worth()
        elif choice == "7":
            filter_by_category()
        elif choice == "8":
            low_stock_alert()
        elif choice == "9":
            restock_item()
        elif choice == "10":
            print("\nüëã Exiting Inventory System. Have a great day!\n")
            break
        else:
            print("‚ùå Invalid choice. Please enter a number from 1 to 10.")

login()
# --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- + --- + --- +  --- + --- + --- + --- + --- + --- + --- + --- +
```
